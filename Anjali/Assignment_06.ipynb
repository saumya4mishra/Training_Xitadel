```python
# 1. You are given two strings s and t. 
# String t is generated by randomly shuffling string s and then adding one more letter at a random position. 
# Return the letter that was added to t.
def find_the_difference(s,t):
    m = [t[i] for i in range(len(t)) if t[i] not in s]
#     d = ''
#     for i in m:
#         d += i
    return m
find_the_difference("abcd", "abcde")
```

```python
# 2. Given a function that accepts unlimited arguments, check and count how many data types are in those arguments. 
# Finally return the total in a list.[int, str, bool, list, tuple, dictionary]
def count_datatypes(*d, **v):
    out = [0, 0, 0, 0, 0, 0]
    for i in d:
        if type(i) == int:
            out[0] += 1
        elif type(i) == str:
            out[1] += 1
        elif type(i) == bool:
            out[2] += 1
        elif type(i) == list:
            out[3] += 1
        elif type(i) == tuple:
            out[4] += 1
        else:
            out[5] += 1
    return out
count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23)
```

```python
# 3. A Fibonacci string is a precedence of the Fibonacci series. 
# It works with any two characters of the English alphabet (as opposed to the numbers 0 and 1 in the Fibonacci series) 
# as the initial items and concatenates them together as it progresses in a similar fashion as the Fibonacci series.
def fib_str(i, s):
    result = ''
    a,b = s[0],s[1]
    if i == 1:
        result += a
        return result
    elif i == 2:
        result += a + ', ' + b
        return result
    else:
        result += a + ', ' + b
        for n in range(3,i+1):
            c = b+a
            a,b = b,c
            result += ', ' + c
    return result
fib_str(6, ["n", "k"])
```

```python
# 5. Fibonacci using recursion
def fib(n):
    if n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fib(n-2) + fib(n-1)
fib(7)
```
